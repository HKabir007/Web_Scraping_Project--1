```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
Cricket_team = []

for i in soup.find_all('span',class_="u-hide-phablet"):
    Cricket_team.append(i.text)
    
Cricket_team
```

```python
Women_cricket_team = Cricket_team[0:10]
Women_cricket_team
```

```python
Matche1 = []

for i in soup.find_all('td',class_="rankings-block__banner--matches"):
    Matche1.append(i.text)
    
Matche1
```

```python
Matches = []

for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    Matches.append(i.text)
    
Matches
```

```python
Match2 = Matche1 + Matches[0::2]
Match2
```

```python
Match = Match2[0:10]
Match
```

```python
Point1 = []

for i in soup.find_all('td',class_="rankings-block__banner--points"):
    Point1.append(i.text)
    
Point1
```

```python
Point2 = Point1 + Matches[1::2]
```

```python
Point = Point2[0:10]
Point
```

```python
Rating1 = []

for i in soup.find_all('td',class_="rankings-block__banner--rating u-text-right"):
    Rating1.append(i.text.replace('\n',''))
    
Rating1
```

```python
Rating2 = []
for rating in Rating1:
    rating = rating.strip(" ")
    Rating2.append(rating)
    
Rating2
```

```python
Ratings = []

for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    Ratings.append(i.text)
    
Ratings
```

```python
Rating = Rating2 + Ratings[0:9]
Rating
```

```python
print(len(Women_cricket_team))
print(len(Match))
print(len(Point))
print(len(Rating))
```

```python
import pandas as pd
df = pd.DataFrame({'Top 10 women cricket teams':Women_cricket_team,'No. of Matches':Match,'Point':Point,'Rating':Rating})
df
```
