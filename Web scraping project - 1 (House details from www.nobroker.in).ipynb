```python
soup = BeautifulSoup(page.content)
soup
```

```python
Heading1 = []

for i in soup.find_all('h2',class_="heading-6 flex items-center font-semi-bold m-0"):
    Heading1.append(i.text)
    
Heading1
```

```python
Location1 = []

for i in soup.find_all('div',class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0 po:max-w-95"):
    Location1.append(i.text)
    
Location1
```

```python
Price1 = []

for i in soup.find_all('div',class_="p-1.5p flex border-b border-b-solid border-cardbordercolor tp:py-1p tp:px-1.5p tp:border-b-0"):
    Price1.append(i.text.replace('Not under loan₹ 1,570 sqftBuiltup',''))
    
Price1
```

```python
Price2 = []
for Price in Price1:
    Price = Price.replace('Under Loan₹ 937 sqftBuiltup','')
    Price = Price.replace('Not under loan₹ 570 sqftBuiltup','')
    Price = Price.replace('Under Loan₹ 775 sqftBuiltup','')
    Price = Price.replace('Not under loan₹ 1,265 sqftBuiltup','')
    Price2.append(Price)
Price2
```

```python
import pandas as pd
df = pd.DataFrame({'House Title':Heading1,'Location':Location1,'EMI & Price':Price2})
df
```
